````python
# Pertama membuat fungsi dulu untuk menjalankan algoritma Palindrom Checker nya
def palindrome(numbers: list) -> bool:
  # Bisa membuat base case nya
  if len(numbers) <= 1:
    return True # Karena jika hanya ada 1 elemen maka sudah pasti bernilai True
  # Melakukan cek elemen pertama apakah == elemen terakhir lalu menggunakan "and", karena True and True = True
  return numbers[0] == numbers[-1] and is_palindrom(numbers[1:-1]) # Hasilnya adalah True and False (Ketika membandingkan 3 : 4) -> False
print(palindrome([1,2,3,4,2,1])) # numbers[1:-1] itu mengambil elemen +1 dari depan dan -1 dari belakang artinya 123421 menghasilkan -> 2342 -> 34 -> False
````

## Membuat kode program:


````python
# Membuat function untuk menjalankan algoritma program
def cariTerbesar(numbers: list[int]) -> int: # Menambahkan anotasi tipe untuk parameter yang di input berdasarkan argumen dan hasil return dari functionnya
  if len(numbers) == 1: # Base case jika panjang list hanya 1
    return numbers[0] # Mereturn langsung, karena hanya ada 1 elemen
  else: # Membuat recursive case
    big_one = cariTerbesar(numbers[1:]) # Membuat variabel yang mereferensikan fungsi itu sendiri dengan argumen index awal + 1
    if numbers[0] > big_one: # Membuat perbandingan dan perkondisian antara elemen pertama list dan hasil return dari recursive functionnya
      return numbers[0] # Mengembalikan elemen pertama list jika kondisi "if" terpenuhi
    else:
      return big_one # Mereturn variabel big_one yang mana akan memulai rekursif nya hingga bascase yakni ketika sudah elemen terakhir [1:] dari 2 elemen terakhir
print(cariTerbesar( [1,2,3,4,5,1])) # Print outputnya
````

## Clean Code nya:


````python
def cariTerbesar(numbers: list[int]) -> int:
  if len(numbers) == 1:
    return numbers[0]
  else:
    big_one = cariTerbesar(numbers[1:])
    if numbers[0] > big_one:
      return numbers[0]
    else:
      return big_one
print(cariTerbesar( [1,2,3,4,5,1]))
````

Dalam kode program yang saya buat, function yang menjalankan algoritma program untuk mencari angka terbesar memiliki _base case dan recursive case_
Berikut rinciannya: <br>
a. _base case_
```
  if len(numbers) == 1:
    return numbers[0]
```
Dalam _base case_ tersebut sebenarnya bisa berlaku dua fungsi utama:
1. Berfungsi sebagai kondisi yang menghentikan rekursif. Yakni ketika user benar-benar hanya menginputkan list dengan panjang elemen 1 ke argumen sebagai parameter function. Maka _base case_ tersebut berfungsi sebagai perkondisian khusus yakni mengembalikan 1 elemen tersebut.
<br> Contoh: function([2]) -> akan otomatis me-return 2
2. Berfungsi sebagai nilai dasar (base case) dari functionnya untuk proses rekursif. Dalam matematika jika ada bentuk f(x) = y, maka _base case_ ini sendiri merupakan x = 0 dalam f(0) yang akan menjadi nilai awal dari fungsi x.

b. _recursive case_
```
  else:
    big_one = cariTerbesar(numbers[1:])  
    if numbers[0] > big_one:
      return numbers[0]
    else:
      return big_one
```
Dalam _recursive case,_ kita mempraktekkan rumus fungsi f(x) dalam aritmatika yang tadi di jelaskan dalam _base case._ Jadi, nilai dari fungsi dasar selalu me return numbers[0] yang nantinya akan di bandingkan dengan elemen selanjutnya dengan pemanggilan fungsi
```
cariTerbesar(numbers[1:])  
```
Rumusnya terletak pada:
```
if numbers[0] > big_one:
      return numbers[0]
    else:
      return big_one
```
Yang akan membandingkan elemen awal masing-masing pemanggilan rekursif terhadap fungsi tersebut.


````python
# Import librarynya
from visualiser.visualiser import Visualiser as vs
from IPython.display import Image
````

````python
# Siapkan algoritma pada function nya
# Tambah dekorator nya
@vs(node_properties_kwargs={"shape":"record", "color":"#f57542", "style":"filled", "fillcolor":"grey"})
def cariTerbesar(numbers: list[int]) -> int:
  if len(numbers) == 1:
    return numbers[0]
  else:
    big_one = cariTerbesar(numbers[1:])
    if numbers[0] > big_one:
      return numbers[0]
    else:
      return big_one
def main():
    # Panggil Functionnya
    print(cariTerbesar([1,2,3,4,5,1]))
    # Save ke file
    vs.make_animation("cekTerbesar.gif", delay=2)

if __name__ == "__main__":
    main()
````
