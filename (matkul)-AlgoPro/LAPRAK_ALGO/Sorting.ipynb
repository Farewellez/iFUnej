# SORTING

Algoritma sorting adalah proses pengurutan suatu kumpulan data atau elemen-elemen dalam urutan tertentu, biasanya dari yang terkecil ke yang terbesar atau sebaliknya. Tujuan utamanya adalah untuk mempermudah pencarian dan analisis data. Algoritma sorting dapat bervariasi mulai dari yang sederhana seperti bubble sort hingga yang kompleks seperti quicksort atau merge sort.


# Bubble Sort

Bubble sort adalah salah satu algoritma pengurutan sederhana yang bekerja dengan cara membandingkan dan menukar elemen-elemen berpasangan dalam sebuah daftar (array) hingga seluruh elemen terurut sesuai dengan urutan yang diinginkan.

Nama "bubble" berasal dari cara algoritma ini bekerja, yaitu dengan "menggelembungkan" elemen-elemen yang lebih kecil atau lebih besar ke atas atau ke bawah daftar.


https___dev-to-uploads.s3.amazonaws.com_uploads_articles_e9pn1te1h99tp8rhwv07.jpg


```python
my_array = [20, 2, 3, -2, 1, 15]
print(my_array)

n = len(my_array)
for i in range(n - 1):
    for j in range(n - 1 - i):
        if my_array[j] > my_array[j + 1]:
            my_array[j], my_array[j + 1] = my_array[j + 1], my_array[j]

print(my_array)
```

```python
my_array = [20,2,3,-2,1,15]
print(my_array)

isSorted = False
n = len(my_array)
while not isSorted:
    isSorted = True
    for i in range(1,n):
        if my_array[i] < my_array[i-1]:
            isSorted = False
            my_array[i], my_array[i-1] = my_array[i-1], my_array[i]

print(my_array)
```

# Selection Sort

Selection Sort adalah algoritma sorting sederhana yang bekerja dengan cara mencari elemen terkecil dalam suatu list/array yang belum diurutkan untuk kemudian ditukar/dipindahkan posisinya ke dalam list/array yang telah diurutkan.

selection-short.png

```python
my_array = [7,2,1,3,5,4]
print(my_array)

for i,value_i in enumerate(my_array):
  index_of_minimum_elemen = i
  for j,value_j in enumerate(my_array):
    if i >= j:
      continue
    if my_array[j] < my_array[index_of_minimum_elemen]:
      index_of_minimum_elemen = j
  if i < index_of_minimum_elemen:
    t = my_array[i]
    my_array[i] = my_array[index_of_minimum_elemen]
    my_array[index_of_minimum_elemen] = t


print(my_array)
```

```python
my_array = [20,2,3,-2,1,15]
print(my_array)

for i in range(len(my_array)):
  idx_angka_max = i
  for j in range(i,len(my_array)):
    if my_array[j] > my_array[idx_angka_max]:
      idx_angka_max = j

  if i < idx_angka_max:
    my_array[i], my_array[idx_angka_max] = my_array[idx_angka_max], my_array[i]


print(my_array)
```

# Insertion Sort

Insertion sort adalah algoritma pengurutan sederhana yang bekerja dengan membagi array yang tidak terurut menjadi dua bagian: bagian yang sudah diurutkan dan bagian yang belum diurutkan. Pada setiap iterasi, algoritma ini memilih satu elemen dari bagian yang belum diurutkan dan memasukkannya ke posisi yang sesuai dalam bagian yang sudah diurutkan, sehingga memperpanjang bagian yang sudah diurutkan. Proses ini diulang sampai semua elemen telah dimasukkan ke dalam bagian yang sudah diurutkan, sehingga array menjadi terurut.

Insertion-sorting.png


```python
my_array = [7,2,1,3,5,4]
print(my_array)

for i,value_i in enumerate(my_array):
  if i == 0:
    continue

  current_element = my_array[i]
  j = i-1
  while j >= 0:
    if current_element >= my_array[j]:
      break
    my_array[j+1] = my_array[j]
    j = j-1

  my_array[j+1] = current_element

print(my_array)
```

```python
my_array = [20,2,3,-2,1,15]

for i in range(1,len(my_array)):
  angka_i = my_array[i]
  j = i-1
  while j >= 0:
    if angka_i >= my_array[j]:
      break
    my_array[j+1] = my_array[j]
    j -= 1

  my_array[j+1] = angka_i

print(my_array)
```

# Merge Sort

 Merge sort adalah salah satu pengurutan dengan metode memecah data kemudian mengolah untuk diselesaikan pada setiap bagian dan menggabungkan kembali sehingga data tersebut berhasil tersusun. Merge sort dalam menyelesaikan pengurutan membutuhkan fungsi rekursif.


 Divide_And_Conquer_8.jpg

```python
def ms(array):
    print('Memecah List', array)
    n = len(array)
    if n < 2:
        return array
    else:
        mid=n//2
        left=array[:mid]
        right=array[mid:]
        ms(left)
        ms(right)
        i=0
        j=0
        k=0
        while i < len (left) and j < len (right):
            if left[i]>right[j]:
                array[k]=left[i]
                i=i+1
            else:
                array[k]=right[j]
                j=j+1
            k=k+1
        while i < len (left):
            array[k]=left[i]
            i=i+1
            k=k+1
        while j < len (right):
            array[k]=right[j]
            j=j+1
            k=k+1
    print('Menggabungkan List', array)
array = [2,5,60,43,27,10,89,17]
ms(array)
```
